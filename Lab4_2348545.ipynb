```python
#importing data
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/Program_4_Dataset.csv')
data.head()
```

```python
#assigning predictor and target variables
X=data[['Temperature']]
Y=data['Pressure']
```

```python
#plotting the same
import matplotlib.pyplot as plt
plt.scatter(data['Temperature'],data['Pressure'])
```

```python
#splitting data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.44,random_state=2)
```

```python
#fitting linear model
from sklearn.linear_model import LinearRegression
reg_Model = LinearRegression()
reg_Model.fit(X_train,Y_train)
```

```python
#finding R2 score of linear model
from sklearn.metrics import r2_score
score = r2_score(Y_test, reg_Model.predict(X_test))
print(score)
```

```python
#plotting the linear model
plt.plot(X_train,reg_Model.predict(X_train),color='r')
plt.scatter(X_train,Y_train)
```

```python
#fitting polynomial model
from sklearn.preprocessing import PolynomialFeatures
p1=PolynomialFeatures(degree=2,include_bias=True)
x_train_p1=p1.fit_transform(X_train)
x_test_p1=p1.transform(X_test)
```

```python
#finding the R2 score
reg_Model1=LinearRegression()
reg_Model1.fit(x_train_p1,Y_train)
y_predict1=reg_Model1.predict(x_test_p1)
score1=r2_score(Y_test,y_predict1)
print(score1)
```

```python
#plotting the same
plt.scatter(X_train,reg_Model1.predict(x_train_p1))
plt.scatter(X_train,Y_train)
```
