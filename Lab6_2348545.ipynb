```python
#creating breast cancer,iris,diabetes datasets
from sklearn.datasets import load_breast_cancer, load_iris, load_diabetes
import pandas as pd
bc_X = pd.DataFrame(data = load_breast_cancer()['data'],
                       columns=load_breast_cancer()['feature_names'])
bc_y = load_breast_cancer()['target']
iris_X = pd.DataFrame(data = load_iris()['data'],
                       columns=load_iris()['feature_names'])
iris_y = load_iris()['target']
diabetes_X = pd.DataFrame(data = load_diabetes()['data'],
                       columns=load_diabetes()['feature_names'])
diabetes_y = load_diabetes()['target']
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
ss = StandardScaler()
bc_X = pd.DataFrame(ss.fit_transform(bc_X)) #scaling the data
bc_X
pca = PCA(n_components=10)
bc_X_final = pd.DataFrame(pca.fit_transform(bc_X))
bc_X_final
```

```python
ss = StandardScaler()
iris_X = pd.DataFrame(ss.fit_transform(iris_X)) #scaling the data
iris_X
pca = PCA(n_components=2)
iris_X_final = pd.DataFrame(pca.fit_transform(iris_X))
iris_X_final
```

```python
ss = StandardScaler()
diabetes_X = pd.DataFrame(ss.fit_transform(diabetes_X)) #scaling the data
diabetes_X

pca = PCA(n_components=7)
db_X_final = pd.DataFrame(pca.fit_transform(diabetes_X))
db_X_final
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
bc_X_final_lda = pd.DataFrame(lda.fit_transform(bc_X,
                                                y=load_breast_cancer()['target']))
bc_X_final_lda
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.scatter(bc_X_final_lda, y=np.zeros(len(bc_X_final_lda)),
            c=load_breast_cancer()['target'])
plt.show()
```

```python
lda = LinearDiscriminantAnalysis()
iris_X_final_lda = pd.DataFrame(lda.fit_transform(iris_X,
                                                y=load_iris()['target']))
iris_X_final_lda
```

```python
plt.scatter(iris_X_final_lda[0], iris_X_final_lda[1],
            c=load_iris()['target'])
plt.show()
```

Since Diabetes dataset has a continuous target variable, this is a regression problem. Therefore, LDA is not applicable on this dataset.

```python
#applying KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
knn_classifier = KNeighborsClassifier(n_neighbors=3)
knn_regressor = KNeighborsRegressor(n_neighbors=3)
```

```python
#breast cancer dataset
knn_classifier.fit(bc_X_final_lda, load_breast_cancer()['target'])
bcd1=knn_classifier.score(bc_X_final_lda, load_breast_cancer()['target'])
print("Mean accuracy:",bcd1)
```

```python
#iris dataset
knn_classifier.fit(iris_X_final_lda, load_iris()['target'])
iris1=knn_classifier.score(iris_X_final_lda, load_iris()['target'])
print("Mean accuracy:",iris1)
```

```python
#diabetes dataset
knn_regressor.fit(db_X_final, load_diabetes()['target'])
dia1=knn_regressor.score(db_X_final, load_diabetes()['target'])
print("R2-score:",dia1)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor

dt_classifier = DecisionTreeClassifier(max_depth=5)
dt_regressor = DecisionTreeRegressor(max_depth=5)
```

```python
#breast cancer dataset
dt_classifier.fit(bc_X_final_lda, load_breast_cancer()['target'])
bcd2=dt_classifier.score(bc_X_final_lda, load_breast_cancer()['target'])
print("Mean accuracy:",bcd2)
```

```python
#iris dataset
dt_classifier.fit(iris_X_final_lda, load_iris()['target'])
iris2=dt_classifier.score(iris_X_final_lda, load_iris()['target'])
print("Mean accuracy:",iris2)
```

```python
#diabetes dataset
dt_regressor.fit(db_X_final, load_diabetes()['target'])
dia2=dt_regressor.score(db_X_final, load_diabetes()['target'])
print("R2-score:",dia2)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor

rf_classifier = RandomForestClassifier(max_depth=5)
rf_regressor = RandomForestRegressor(max_depth=5)
```

```python
#breast cancer dataset
rf_classifier.fit(bc_X_final_lda, load_breast_cancer()['target'])
bcd3=rf_classifier.score(bc_X_final_lda, load_breast_cancer()['target'])
print("Mean accuracy:",bcd3)
```

```python
#iris dataset
rf_classifier.fit(iris_X_final_lda, load_iris()['target'])
iris3=rf_classifier.score(iris_X_final_lda, load_iris()['target'])
print("Mean accuracy:",iris3)
```

```python
#diabetes dataset
rf_regressor.fit(db_X_final, load_diabetes()['target'])
dia3=rf_regressor.score(db_X_final, load_diabetes()['target'])
print("R2-score:",dia3)
```

```python
#performance analysis for breast cancer dataset
Y=[bcd1,bcd2,bcd3]
X=[f"KNN ({bcd1:.2f})",f"Decision Tree ({bcd2:.2f})",f"Random forest ({bcd3:.2f})"]
plt.bar(X,Y)
```

Clearly, LDA does an excellent job at separating the classes and it is therefore a relatively easy task to create a decision boundary to differentiate between the classes in the projected dimension.

Best Models are Decision Tree and Random Forest

```python
#performance analysis for iris dataset
Y=[iris1,iris2,iris3]
X=[f"KNN ({iris1:.2f})",f"Decision Tree ({iris2:.2f})",f"Random forest ({iris3:.2f})"]
plt.bar(X,Y)
```

The dataset is perfectly linearly separable which results in all the models performing in an excellent manner.

Best Models are Decision Tree and Random Forest

```python
#performance analysis for diabetes dataset
Y=[dia1,dia2,dia3]
X=[f"KNN ({dia1:.2f})",f"Decision Tree ({dia2:.2f})",f"Random forest ({dia3:.2f})"]
plt.bar(X,Y)
```

Since this is a regression problem, LDA is not applicable in this case. Some other methods have to be used in order to train an optimal model.

Best Model is Random Forest
